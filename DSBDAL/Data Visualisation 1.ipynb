```python
import numpy as np
import pandas as pd
from tabulate import tabulate
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
sns.set_style('whitegrid')
#PLOT 1 - Countplot for the number of passengers in each class
plt.figure(figsize=(8,6))
sns.countplot(x="class", data=titanic,palette='Set1')
plt.title("Passenger Count by Class")
plt.show()
```

```python
#Distribution of Ages using a Histogram
plt.figure(figsize=(8,6))
sns.histplot(titanic['age'].dropna(), bins=40,kde=True, color='skyblue')
plt.title('Distribution of Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
titanic['age'].fillna(titanic['age'].mean())
```
