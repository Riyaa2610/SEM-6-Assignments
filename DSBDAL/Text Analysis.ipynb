```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
nltk.download('omw-1.4')
```

```python
with open('doc_01.txt','r') as file:
    doc = file.read()
#replace non-alphanumeric chars and underscores with whitespaces
doc = re.sub("[\W_]+"," ",doc)
```

```python
#Tokenisation

word_tokens = nltk.word_tokenize(doc)
word_tokens
```

```python
#Stop words

stop_words = set(nltk.corpus.stopwords.words('english'))
word_tokens = [token for token in word_tokens if token not in stop_words]
word_tokens
```

```python
#POS Tagging

tagged = nltk.pos_tag(word_tokens)
tagged
```

```python
#Stemming and Lemmatization

#Lemmatization
lemmatizer = nltk.stem.WordNetLemmatizer()
lemmatized_tokens = [lemmatizer.lemmatize(token) for token in word_tokens]
lemmatized_tokens
```

```python
#Stemming

stemmer = nltk.stem.PorterStemmer()
stemmed_tokens = [stemmer.stem(token) for token in word_tokens]
stemmed_tokens
```

```python
#PART 2
#TF-IDF

def calculate_term_freq(doc):
    word_tokens = nltk.word_tokenize(doc)
    tf_dict = dict()
    for word in word_tokens:
        tf_dict[word] = word_tokens.count(word)
    tf = dict()
    for word , count in tf_dict.items():
        tf[word] = count/len(tf_dict)
    return tf
```

```python
tf = calculate_term_freq(doc)
```

```python
all_tokens = nltk.word_tokenize(doc)

def calculate_idf():
    N = 1
    idf = dict()
    for token in all_tokens:
        f = 1
        if token in all_tokens:
            f+=1
        idf[token] = math.log(N/f)
    return idf
```

```python
idf = calculate_idf()
```

```python
d_repr = []
for token in all_tokens:
    d_repr.append(tf[token]*idf[token])
d_repr
```
